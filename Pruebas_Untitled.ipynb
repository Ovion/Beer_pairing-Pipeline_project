```python
def to_drop(df, lst, indx):
    return df.drop(lst, inplace = True, axis = indx)

lst_colmn = ['abv', 'ibu', 'Unnamed: 0']
to_drop(df_beers, lst_colmn, 1)
df_beers.head(10)
```

```python
df_beers['style'].fillna('*', inplace=True)

lst_row = [566, 1554, 1555]
to_drop(df_beers, lst_row, 0)
df_beers.loc[df_beers['style'] == '*']
```
